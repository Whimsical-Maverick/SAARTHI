```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
import os
import pandas as pd

def merge_csvs_in_directory(directory_path):
  all_dataframes = []
  for filename in os.listdir(directory_path):
    if filename.endswith(".csv"):
      filepath = os.path.join(directory_path, filename)
      try:
        df = pd.read_csv(filepath)
        all_dataframes.append(df)
      except Exception as e:
        print(f"Error reading file {filepath}: {e}")

  if not all_dataframes:
    return pd.DataFrame()  # Return empty DataFrame if no CSV files found

  merged_dataframe = pd.concat(all_dataframes, ignore_index=True)
  return merged_dataframe
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score
```

```python
scaler=MinMaxScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled=pd.DataFrame(X_scaled,columns=X.columns)
```

```python
X_scaled
```

```python
for col in X_scaled.columns:
  sns.boxplot(X_scaled[col])
  plt.title(col)
  plt.show()
```

```python
def get_number_of_Clusters(K,inertia):
  mapping = {}
  for k in K:
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    mapping[k]=kmeans.inertia_
  return mapping
```

```python
def label_data_by_clusters(df,features,feature_df,Fit):
  kmeans=KMeans(n_clusters=Fit,init='k-means++',random_state=42)
  kmeans.fit(feature_df)
  df[features]=feature_df
  df['cluster']=kmeans.labels_
  return df,kmeans.cluster_centers_
```

```python
def plot_elbow(K,inertia):
  plt.figure(figsize=(16,8))
  plt.plot(K,inertia,'bx-')
  plt.xlabel('Number of clusters')
  plt.ylabel('Inertia')
  plt.title('Elbow Method')
  plt.show()
```

```python
Range=range(1,11)
inertia=[]
mapping=get_number_of_Clusters(Range,inertia)
```

```python
print(mapping)
```

```python
plot_elbow(Range,inertia)
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
kmeans.fit(X_scaled)
```

```python
score=silhouette_score(X_scaled,kmeans.labels_)
print(score)
```

```python
data,centroids=label_data_by_clusters(data,features,X,4)
```
